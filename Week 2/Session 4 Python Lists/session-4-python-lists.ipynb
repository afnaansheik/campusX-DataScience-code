## Accessing Items from a 
-  Indexing
- Slicing

```python
# Indexing
L = [1,2,3,4,5]

# Positive Indexing  
print(L[0])

# Negative Indexing  
print(L[-1])
```

```python
L = [1,2,3,[4,5]]

# Positive Indexing
print(L[3][0])
# Negative Indexing
print(L[-1][-2])
```

## Operations on Lists

- Arithmetic

```python
# Arithmetic (+, *)

L1 = [1,2,3,4]
L2 = [5,6,7,8]

# concatenation/merge
print(L1+L2)
```

## List Functions

```python
# len/min/max/sorted
L = [2,1,5,7,0]

print(len(L))

print(min(L))
print(max(L))

print(sorted(L))
print(sorted(L,reverse=True))
```

```python
# Index
L = [1,2,1,3,4,1,5]

L.index(2)
```

```python
# sort (vs sorted)
L = [2,1,5,7,0]
print(L)
print(sorted(L)) # create new memory & sort but dnt change in originall list
print(L)
L.sort() # permently sort 
print(L)
```

```python
# copy -> shallow 
L = [2,1,5,7,0]
print(L)
print(id(L))
L1 = L.copy()
print(L1)
print(id(L1))
```

## List Comprehension

List Comprehension provides a concise way of creating lists.



Advantages of List Comprehension
- More time-efficient and space-efficient than loops.
- Require fewer lines of code.
- Transforms iterative statement into a formula.

```python
# Add 1 to 10 numbers to a list
# we have to write 4 lines for this code
L = []
for i in range(1,11):
    L.append(i)
    
print(L)
```

```python
# List Comprehension can take 1 line code
L = [i for i in range(1,11)]
print(L)
```

```python
# scalar multiplication on a vector
v = [2,3,4]
s = 3

x = []
for i in v:
    x.append(i*s)

print(x)
```

```python
# List Comprehension using scalar multiplication on a vector
v = [2,3,4]
s = 3

[s*i for i in v]
```

```python
# Add squares
L = [1,2,3,4,5]

sq = []
for i in L:
    sq.append(i*i)

print(sq)
```

```python
# List Comprehension using Add squares
L = [1,2,3,4,5]

[i*i for i in L]
```

```python
# Print all numbers divisible by 5 in the range of 1 to 50
L = []

for i in range(1,51):
    L.append(i)

# print(L)

for i in L:
    if (i%5==0):
        print(i)
```

```python
# # List Comprehension using Print all numbers divisible by 5 in the range of 1 to 50
[i for i in range(1,51) if i%5==0]
```

```python
# find languages which start with letter p
languages = ['java','python','php','c','javascript']

[languages for languages in languages  if languages.startswith('p') ]
```

```python
# Nested if with List Comprehension
basket = ['apple','guava','cherry','banana']
my_fruits = ['apple','kiwi','grapes','banana']

# add new list from my_fruits and items if the fruit exists in basket and also starts with 'a'

[fruit for fruit in my_fruits if fruit in basket if fruit.startswith('a')]
```

```python
# Print a (3,3) matrix using list comprehension -> Nested List comprehension
[[i*j for i in range(1,4)] for j in range(1,4)]
```

```python
# cartesian products -> List comprehension on 2 lists together
L1 = [1,2,3,4]
L2 = [5,6,7,8]

[i*j for i in L1 for j in  L2]
```

## 2 ways to traverse a list

- itemwise
- indexwise

```python
# itemwise
L1 = [1,2,3,4]

for i in L1:
    print(i)
```

```python
# indexwise
L1 = [1,2,3,4]

for i in range(0,len(L1)):
    print(i)
```

## Zip

The zip() function returns a zip object, which is an iterator of tuples where the first item in each passed iterator is paired together, and then the second item in each passed iterator are paired together.

If the passed iterators have different lengths, the iterator with the least items decides the length of the new iterator.

```python
# Write a program to add items of 2 lists indexwise
L1 = [1,2,3,4]
L2 = [5,6,7,8]

print(zip(L1,L2))
print(list(zip(L1,L2)))

[i+j for i,j in zip(L1,L2)]
```

- Risky usage

```python
# Risky usage
a = [1,2,3,4]
b = a

print(a)
print(b)

a.append(4) # if i change value on a it change update value in b bcz it direct that address point

print(a)
print(b)
```
